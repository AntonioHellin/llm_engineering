Para demostrar que estás familiarizado con la API de OpenAI y también con Ollama, crea una herramienta que responda a una pregunta técnica
y la explique. ¡Esta es una herramienta que podrás usar durante el curso!

```python
# imports

from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
```

```python
# constantes

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# Configura el entorno

load_dotenv()
openai = OpenAI()
```

```python
# Aquí está la pregunta; escribe sobre esto para preguntar algo nuevo

question = input("Por favor, introduce tu pregunta técnica: ")
```

```python
# Haz que gpt-4o-mini responda, con streaming

# 1. Preparamos las instrucciones (el rol del sistema y la pregunta del usuario)
messages = [
    {"role": "system", "content": "Eres un tutor técnico experto en programación."},
    {"role": "user", "content": question}
]

# 2. Hacemos la llamada a OpenAI
stream = openai.chat.completions.create(
    model=MODEL_GPT, 
    messages=messages,
    stream=True
)

# 3. Mostramos la respuesta progresivamente
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    texto = chunk.choices[0].delta.content or ''
    response += texto
    update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Haz que Llama 3.2 responda
respuesta_ollama = ollama.chat(
    model=MODEL_LLAMA, 
    messages=messages
)

# Extraemos el contenido del JSON de respuesta y lo mostramos
texto_final = respuesta_ollama['message']['content']
display(Markdown(texto_final))
```
